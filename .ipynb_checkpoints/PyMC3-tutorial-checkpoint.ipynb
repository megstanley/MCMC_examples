# How to use PyMC3

Perhaps the most useful part of this begins with the example under the heading 'Stochastic Volatility'

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')

# Initialize random number generator
np.random.seed(123)

# True parameter values
alpha, sigma = 1, 1
beta = [1, 2.5]

# Size of dataset
size = 100

# Predictor variable
X1 = np.random.randn(size)
X2 = np.random.randn(size) * 0.2

# Simulate outcome variable
Y = alpha + beta[0]*X1 + beta[1]*X2 + np.random.randn(size)*sigma
```

Here we are looking at a Bayesian normal regression model. We have an expected distribution, from which the samples are assumed to be drawn. The parameters of the distribution are $\mu$ and $\sigma$, and $\mu$ is actually a linear combination of the features, $\mu$ = $\alpha$ + $\beta_{i}X_{i}$. For the parameters $\alpha$ and $\beta$ we will assume a zero mean high variance prior (weak information, basically uniform) and for the parameter $\sigma$ we choose a single sided normal with variance 1.

We have already simulated false data above. Let's plot it.

```python
%matplotlib inline

fig, axes = plt.subplots(1, 2, sharex=True, figsize=(10,4))
axes[0].scatter(X1, Y)
axes[1].scatter(X2, Y)
axes[0].set_ylabel('Y'); axes[0].set_xlabel('X1'); axes[1].set_xlabel('X2');
```

The values of X are normally distributed by design, and this leads to the values of Y, linearly related, appearing as though they may be too. But the question is -- how can we correctly infer the parameters of the distribution of Y?

# Using PyMC3

```python
import pymc3 as pm
print('Running on PyMC3 v{}'.format(pm.__version__))
```

```python
basic_model = pm.Model()

with basic_model:

    # Priors for unknown model parameters
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=10, shape=2)
    sigma = pm.HalfNormal('sigma', sd=1)

    # Expected value of outcome
    mu = alpha + beta[0]*X1 + beta[1]*X2

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=Y)
```

```python
with basic_model:
    # draw 500 posterior samples
    trace = pm.sample(500)
```

```python
pm.traceplot(trace);
```

```python
pm.summary(trace).round(2)
```
