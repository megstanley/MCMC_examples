# Using emcee package for MCMC -- an example

Let's do some Bayesian inference while we work on a super simple model, to learn how to use emcee package.

I am going to generate some synthetic data for the simple generative model $y = ax + b$ with some noise added on the data. 

Eventually, we want to use this to infer distributions of parameters that are most likely given our data and the forward spectral graph model.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
import emcee
import corner
import scipy.optimize as op
```

```python
N = 50

#some 'true' parameters to generate the data
a_true = -0.9594
b_true = 4.294
frac_t = 0.534

x = np.sort(10*np.random.rand(N))
y = a_true*x + b_true

#now we make an assumption about the measurement error in our data. We are assuming that we have errors of zero mean, 
#fractional standard deviation from our data points, which we model by picking random points from a normal distribution 
#where we scale to reflect that size at each data point. 
y += frac_t*y *np.random.rand(N)

#However, we actually want these errors to be underestimates of the true error in the data, so we add some additional 
#noise.
yerr = 0.1+0.5*np.random.rand(N) #additional variance we give to the data.
y += yerr*np.random.rand(N)
```

```python
plt.errorbar(x,y, yerr, fmt='o')
ax = plt.gca()
ax.set_xlabel('x')
ax.set_ylabel('y_measured')
ax.set_ylim(-8,8)
```

Alright, in order to use a MCMC method, we look to maximise the likelihood. That is, we have our data $D$, we have some model with parameters $\Theta$ and we want to maximise $P(D|\Theta)$. This is not everything in Bayesian inference, but let's start here. 

Actually, it is better to use log likelihood -- features a sum over the data points instead of a multplication of the likelihood of all points together. Remember, we are assuming that all data points are identical, and that the errors are independent, and normally distributed with zero mean. 

Let's define a function that calculates $lnP(D|\Theta)$, where the Gaussian distribution for the errors we choose has a variance $s_{n}^{2}$ for the $nth$ data point. We know, because of the way the data has been generated, that this is in fact a sum of some underlying assumed error which underestimates the error, plus additional noise with variance $\sigma_{n}^{2}$. 

$$
s_{n}^{2} = \sigma_{n}^{2} + f^{2}*(ax + b)^{2}
$$

$$
lnP(y_{n}|a, b,f) = -\frac{1}{2}\sum_{n=1}^{N}[{\frac{(y_{n} - ax_{n} - b)^{2}}{s_{n}^{2}}} + ln(2\pi s^{2}_{n})]
$$

Let's make a function that calculates this for us:

```python
def ln_likelihood(theta, x, y, yerr):
    a, b, lnf = theta
    model = (a*x + b)
    var = yerr**2 + (np.exp(2*lnf)) * model**2
    invar = 1.0/(yerr**2 + (np.exp(2*lnf)) * model**2)
    return -0.5*(np.sum(((y-model)**2)*invar + np.log(var)))

#(here I am just writing the function in a slightly different way because it was not agreeing with the example.)
def lnlike(theta, x, y, yerr):
    m, b, lnf = theta
    model = m * x + b
    inv_sigma2 = 1.0/(yerr**2 + (model**2)*np.exp(2*lnf))
    return -0.5*(np.sum((y-model)**2*inv_sigma2 - np.log(inv_sigma2)))
```

OK, that makes sense. Let's find the optimum using scipy.

```python
#use lambda to rename the likelihood for passing in to the optimizer (just for the sake of understanding how that works)
nll = lambda *args: -lnlike(*args)
result = op.minimize(nll, [a_true, b_true, np.log(frac_t)], args=(x, y, yerr))
a_ml, b_ml, lnf_ml = result["x"]
```

## MCMC

We first need to define the prior distributions for the parameters. In light of 'knowing no better', I am going to pick uninformative priors defined over a large range surrounding those that were used to generate the data in the first place. 

So, I am going to say 

```python
def lnpriors(theta):
    a, b, lnf = theta
    if -5.0 < a < 0.5 and 0.0 < b < 10.0 and -10.0 < lnf < 1.0:
        return 0.0
    return -np.inf
#This just says that if the parameters are out of range, then the lnPrior is negative infinity, and otherwise it is zero
#up to a constant (because we are doing ln we can add, not multiply, so this becomes irrelevant for optimization).
```

```python
def lnprobs(theta, x, y, yerr):
    lp = lnpriors(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

Let's initialise the random walkers in the parameter space. For now, I am going to use the 'optimized' values plus some random value draw from a normal distribution. The space dimension is 3. 

```python
ndim, nwalkers = 3, 100
```

```python
pos = [result['x'] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)] #starting position in parameter space.
```

## Running emcee

Set up the sampler, and then run it.

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprobs, args=(x, y, yerr))
```

```python
sampler.run_mcmc(pos, 5000)
```

```python
type(sampler.chain)
```

```python
sampler.chain.shape
```

OK, an array, where for each time point for each walker the value of the parameter is recorded.

```python
time = np.arange(1,5001)
for i in range(sampler.chain.shape[0]):
    plt.plot(time, sampler.chain[i,:,0], linewidth = 0.1, color = 'black')

ax3 = plt.gca()
ax3.set_xlabel('step')
ax3.set_ylabel('a')
```

Let's do the same for the other parameters

```python
time = np.arange(1,5001)
for i in range(sampler.chain.shape[0]):
    plt.plot(time, sampler.chain[i,:,1], linewidth = 0.1, color = 'blue')

ax4 = plt.gca()
ax4.set_xlabel('step')
ax4.set_ylabel('b')
```

```python
for i in range(sampler.chain.shape[0]):
    plt.plot(time, sampler.chain[i,:,2], linewidth = 0.1, color = 'skyblue')

ax5 = plt.gca()
ax5.set_xlabel('step')
ax5.set_ylabel('frac')
```

It is clear from just looking at the chains that the 'burn in' time is approximately 100 steps. For now, let's just take everything after 1000.

So, let's take just those values, and put all the walker values together in one long vector for each parameter. 

```python
values = sampler.chain[:,1000:,:]
```

```python
values.shape
```

```python
values = values.reshape((-1,ndim))
```

There happens to be a lovely module called 'corner' that we can use to make plots of this, thanks to the guys who made emcee!

```python
fig = corner.corner(values, labels=["$a$", "$b$", "$\ln\,frac$"],
                      truths=[a_true, b_true, np.log(frac_t)])
```

```python
xl = np.array([0, 10])
for a, b, lnf in values[np.random.randint(len(values), size=100)]:
    plt.plot(xl, a*xl+b, color="k", alpha=0.1)
plt.plot(xl, a_true*xl+b_true, color="r", lw=2, alpha=0.8)
plt.errorbar(x, y, yerr=yerr, fmt=".k")
```

This looks pretty good! The generative model is no longer a really good way of getting the parameters with all the extra noise on top, and if we take values from the distributions we have inferred, we definitely get something more descriptive of the data.
